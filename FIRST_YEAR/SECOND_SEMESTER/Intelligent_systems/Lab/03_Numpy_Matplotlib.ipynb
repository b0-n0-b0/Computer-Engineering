We should get familiar with the concept of **tensor**: 
it is a generalization of vector and matrices, with an arbitrary number of dimensions.

```python
# define a function that prints all properties of a tensor
def print_properties(my_tensor):
  print('value')
  print(my_tensor)
  print('properties')
  print(f'\tshape: {my_tensor.shape}') 
  print(f'\tnumber of axis: {my_tensor.ndim}')
  print(f'\ttype of elements: {my_tensor.dtype}')
```

```python
# scalar: 0D tensor
my_scalar = np.array(7)
print_properties(my_scalar)
```

```python
# vector: 1 axis - 1D tensor
list_of_numbers = [2,3,5,6,1,-23]
my_vector = np.array(list_of_numbers)
print_properties(my_vector)
```

```python
# matrices: 2 axes - 2D tensor
my_matrix = np.random.random((4,3))
print_properties(my_matrix)
```

## Tensor Operations
What if i want to add a constant number to a 2D tensor (**element-wise** addition)?

```python
def naive_add(): 
  x = np.random.rand(100,100)
  y = np.ones((100,100))*2
  assert len(x.shape) == 2 # x is a 2D Numpy tensor.
  assert x.shape == y.shape
  z = np.zeros(x.shape)
  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      z[i, j] = x[i, j] + y[i, j]
  return z

def numpy_add(): # numpy already has its own implementation of addition (np.add)
  # low level, highly parallel, computationally efficient routines (typically implemented in C)
  x = np.random.rand(100,100)
  y = np.ones((100,100))*2
  return np.add(x,y)

# when there is no ambiguity, numpy implementation supports broadcast:
# the smaller array is "broadcast" across the larger array so that they
# have compatible shapes. Broadcasting provides a means of vectorizing 
# array operations so that looping occurs in C instead of Python.
def broadcast_add(): 
  x = np.random.rand(100,100)
  z = 2
  return np.add(x,z)

print(naive_add()[:3,:3])
# computationally ineffìcient
# memory inefficient

print(numpy_add()[:3,:3])
# computationally effìcient
# memory inefficient

print(broadcast_add()[:3,:3])
# computationally efficient
# memory efficient
```

Runtime evaluation

```python
import timeit

print('NAIVE ADD (s):')
naive_add_t = timeit.timeit('naive_add()', number=10, setup="from __main__ import naive_add")
print(naive_add_t)

print('NUMPY ADD (s):')
numpy_add_t = timeit.timeit('numpy_add()', number=10, setup="from __main__ import numpy_add")
print(numpy_add_t)

print('BROADCAST ADD (s):')
broadcast_add_t = timeit.timeit('broadcast_add()', number=10, setup="from __main__ import broadcast_add")
print(broadcast_add_t)

%timeit naive_add()
%timeit numpy_add()
%timeit broadcast_add()
```

A common tensor operation: **dot product**.


> Note that unlike MATLAB, `*` is elementwise multiplication, not matrix multiplication. We instead use the dot function to compute inner products of vectors, to multiply a vector by a matrix, and to multiply matrices.



```python
import numpy as np
x = np.random.rand(2,3)
y = np.random.rand(3,2)
z = np.dot(x,y)

# matrix multiplication
for tensor in [x,y,z]:
  print(tensor.shape)
  print(tensor)

# scalar product
a = np.array([1,2,3])
b = np.array([100,10,1])
c = np.dot(a,b)
print(c)
```

```python
t = np.arange(0,10,0.1)
sin_t = np.sin(2*np.pi*t/5)
cos_t = np.cos(2*np.pi*t/5)

plt.plot(t,sin_t)
plt.title('Sin function')
plt.xlabel('time')
plt.ylabel('values')
plt.show()
```

```python
import os
if not os.path.exists("img"):
  os.makedirs("img")
```

```python
plt.plot(t,cos_t,'.r', label = 'cosine')
plt.plot(t,sin_t,'--g', label = 'sin')
plt.title('two functions')
plt.xlabel('time')
plt.ylabel('values')
plt.ylim([-1.1,1.5])
plt.legend(loc='upper center')

plt.savefig('img/example_figure.png', format = 'png')
plt.show()
```

```python
# Two subplots, the axes array is 1-d
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(t,cos_t,'.r')
axarr[1].plot(t,sin_t,'.b')
plt.tight_layout()
plt.show()
```

```python
# Two subplots, the axes array is 1-d
f, axarr = plt.subplots(nrows = 1, ncols = 2, sharey=True)
axarr[0].plot(t,cos_t,'.r')
axarr[1].plot(t,sin_t,'.b')
plt.tight_layout()
plt.show()
```
